```python
from tqdm.auto import tqdm
import couchdb
import json
import pprint
import operator
```

```python
couch = couchdb.Server('http://admin:12345@127.0.0.1:5984')

dbmelb14raw = couch['melb14raw']
dbmelb14food = couch['melb14food']
```

```python
jsf = open("menuitems.json")
menudata = json.load(jsf)

foodlist = menudata["menuItems"]
lcfoodlist = [x.lower() for x in foodlist]
```

```python
pp = pprint.PrettyPrinter(indent=4)

# filter "gluttony tweet" and save them to another database
cnt = 0
for i in tqdm(dbmelb14raw):
    
    d = dbmelb14raw[i]
    key = "full_text"
    if key in d:
        t_text = d[key]
        t_textc = t_text.lower()

        containfood = any(substring in t_textc for substring in lcfoodlist)
        #[s for s in lcfoodlist if s in t_textc]
        
        if containfood:
            newdic = {"foodtags" : [s for s in lcfoodlist if s in t_textc]}
            #print( [s for s in lcfoodlist if s in t_textc] )
            d.update(newdic)
            d.pop("_id")
            d.pop("_rev")
#             pp.pprint(d)
            
            # also should store the food tags as an attribute
            # find the food 
            dbmelb14food.save(d)
            

print("done")
        
```

```python
dat = dbmelb14food["_design/foodtags"]['views']['new-view2']

user_food_cnt = {}
dv = dbmelb14food.view('_design/foodtags/_view/new-view', reduce=True, group=True)
for row in dv:
#     print(row.key, row.value)
    user_food_cnt[row.key] = row.value


sorted_food_cnt = sorted(user_food_cnt.items(), key=operator.itemgetter(1), reverse=True)
pp.pprint(sorted_food_cnt)
```
